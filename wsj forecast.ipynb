```python
import matplotlib.pyplot as plt
import numpy as np
```

# WSJ historical data (2018-2022)

from https://www.wsj.com/articles/u-s-ev-sales-jolted-higher-in-2022-as-newcomers-target-tesla-11672981834

plot title: "Electric vehicle share of total new vehicles sold in the U.S."

```python
# calculate absolute values based on relative heights of bars in graph
wsj_hist = np.array([61, 65, 84, 154, 271])
value_2022 = 5.8
pcts = wsj_hist / wsj_hist[-1] * value_2022
```

```python
years = np.arange(2018, 2023)
plt.bar(x=years, height=pcts)
plt.ylim(top=7)
plt.ylabel("percent")
plt.title("EV share of total new vehicles sold in the U.S.")

for yr, h in zip(years, pcts):
    plt.text(x=yr, y=h+0.3, s=f"{round(h, 1)}%", ha='center', va='bottom', fontsize=12)
```

```python
x = np.arange(2018, 2023)
y = np.round(pcts, 1)
plt.scatter(x, y)

xx = np.arange(2018, 2030, 0.01)

# linear
model1 = np.poly1d(np.polyfit(x, y, 1))
plt.plot(xx, model1(xx), c='red', label="linear")

pred1 = model1(2030)
plt.plot(2030, pred1, marker='o', c='red')
plt.text(2031, pred1, s=f"{round(pred1, 2)}%", c='red')

# quadratic
model2 = np.poly1d(np.polyfit(x, y, 2))
plt.plot(xx, model2(xx), c='green', label="quadratic")

pred2 = model2(2030)
plt.plot(2030, pred2, marker='o', c='green')
plt.text(2031, pred2, s=f"{round(pred2, 2)}%", c='green')

# exponential
# model = np.poly1d(np.polyfit(x, np.log(y), 1))
# plt.plot(future, np.exp(model(future)), label="exponential")

plt.legend()
plt.title("Forecasted EV share of US new vehicles")

plt.show()
```
